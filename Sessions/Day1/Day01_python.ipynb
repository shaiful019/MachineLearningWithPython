```
# prompt: write a python code to get data from api then use matplotlib to print graph  (chose any open source link ) like count some number of things in something
import requests
import matplotlib.pyplot as plt

# Replace with the actual API endpoint you want to use
api_url = 'https://rickandmortyapi.com/api/character'

# Fetch data from the API
response = requests.get(api_url)
data = response.json()

# Extract the number of characters per species (replace with your specific logic)
species_counts = {}
for character in data['results']:
  species = character['species']
  if species in species_counts:
    species_counts[species] += 1
  else:
    species_counts[species] = 1

# Create a bar chart to display the data
species = list(species_counts.keys())
counts = list(species_counts.values())

plt.bar(species, counts)
plt.xlabel('Species')
plt.ylabel('Number of Characters')
plt.title('Number of Characters per Species')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Variable**

A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume).

Rules for Python variables:


*   A variable name must start with a letter or the underscore character
*   A variable name cannot start with a number
*   A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
*   Variable names are case-sensitive (age, Age and AGE are three different variables)
